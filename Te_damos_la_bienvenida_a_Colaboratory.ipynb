```
import matplotlib.pyplot as plt
import numpy as np

# Datos ficticios para los 6 sprints
sprints = np.arange(1, 7)
tasa_exito = [60, 65, 70, 75, 80, 85]
errores_validacion = [10, 8, 5, 3, 2, 1]
layout_diseno = [5, 6, 7, 8, 9, 9]
funcionalidad = [6, 7, 8, 8, 9, 9]
compatibilidad = [5, 6, 7, 8, 8, 9]
documentacion_resultados = [7, 7, 8, 8, 9, 10]
satisfaccion_usuario = [6, 7, 8, 8, 9, 9]

plt.figure(figsize=(14, 10))

# Gráfico de la Tasa de Éxito de Envío
plt.subplot(2, 3, 1)
plt.plot(sprints, tasa_exito, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Tasa de Éxito de Envío (%)')
plt.xlabel('Sprint')
plt.ylabel('Tasa de Éxito (%)')
plt.ylim(50, 90)
plt.grid(True)

# Gráfico del Número de Errores de Validación
plt.subplot(2, 3, 2)
plt.plot(sprints, errores_validacion, marker='o', color='r', linestyle='-', linewidth=2, markersize=8)
plt.title('Número de Errores de Validación')
plt.xlabel('Sprint')
plt.ylabel('Número de Errores')
plt.ylim(0, 12)
plt.grid(True)

# Gráfico de Verificación de Layout y Diseño
plt.subplot(2, 3, 3)
plt.plot(sprints, layout_diseno, marker='o', color='g', linestyle='-', linewidth=2, markersize=8)
plt.title('Verificación de Layout y Diseño')
plt.xlabel('Sprint')
plt.ylabel('Puntaje (0-10)')
plt.ylim(0, 10)
plt.grid(True)

# Gráfico de Verificación de Funcionalidad
plt.subplot(2, 3, 4)
plt.plot(sprints, funcionalidad, marker='o', color='m', linestyle='-', linewidth=2, markersize=8)
plt.title('Verificación de Funcionalidad')
plt.xlabel('Sprint')
plt.ylabel('Puntaje (0-10)')
plt.ylim(0, 10)
plt.grid(True)

# Gráfico de Verificación de Compatibilidad
plt.subplot(2, 3, 5)
plt.plot(sprints, compatibilidad, marker='o', color='c', linestyle='-', linewidth=2, markersize=8)
plt.title('Verificación de Compatibilidad')
plt.xlabel('Sprint')
plt.ylabel('Puntaje (0-10)')
plt.ylim(0, 10)
plt.grid(True)

# Gráfico de Documentación de Resultados
plt.subplot(2, 3, 6)
plt.plot(sprints, documentacion_resultados, marker='o', color='y', linestyle='-', linewidth=2, markersize=8)
plt.title('Documentación de Resultados')
plt.xlabel('Sprint')
plt.ylabel('Puntaje (0-10)')
plt.ylim(0, 10)
plt.grid(True)

plt.tight_layout()
plt.show()
```
